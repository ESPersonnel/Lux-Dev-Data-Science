```python
import matplotlib.pyplot as plt

def logistic_map(r, x):
    return r * x * (1 - x)

def plot_logistic_map(r, x0, n):
    x = [x0]
    for i in range(n-1):
        x.append(logistic_map(r, x[-1]))
    
    plt.plot(x, 'o-')
    plt.xlabel('Iteration')
    plt.ylabel('x')
    plt.title(f'Logistic Map (r = {r}, x0 = {x0})')
    plt.show()

# Example usage:
plot_logistic_map(r=3.6, x0=0.5, n=100)
```

```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

def logistic_map(r, x):
    return r * x * (1 - x)

def plot_logistic_map(r, x0, n):
    x = [x0]
    for i in range(n-1):
        x.append(logistic_map(r, x[-1]))
    
    plt.plot(x, 'o-')
    plt.xlabel('Iteration')
    plt.ylabel('x')
    plt.title(f'Logistic Map (r = {r:.2f}, x0 = {x0:.2f})')
    plt.show()

# Define sliders for r and x0
r_slider = widgets.FloatSlider(value=3.5, min=0, max=4, step=0.01, description='r:')
x0_slider = widgets.FloatSlider(value=0.5, min=0, max=1, step=0.01, description='x0:')

# Define plot output widget
plot_output = widgets.Output()

# Define callback function for sliders
def update_plot(*args):
    plot_output.clear_output()
    with plot_output:
        plot_logistic_map(r_slider.value, x0_slider.value, n=100)

# Connect sliders to callback function
r_slider.observe(update_plot, 'value')
x0_slider.observe(update_plot, 'value')

# Display sliders and plot output
display(widgets.VBox([r_slider, x0_slider, plot_output]))

# Initial plot
plot_logistic_map(r_slider.value, x0_slider.value, n=100)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, max_iter=100):
    z = 0
    for i in range(max_iter):
        z = z**2 + c
        if abs(z) > 2:
            return i
    return max_iter

def generate_mandelbrot_set(xmin, xmax, ymin, ymax, nx=800, ny=800, max_iter=100):
    x = np.linspace(xmin, xmax, nx)
    y = np.linspace(ymin, ymax, ny)
    mandelbrot_set = np.zeros((ny, nx))

    for i in range(ny):
        for j in range(nx):
            c = x[j] + y[i]*1j
            mandelbrot_set[i,j] = mandelbrot(c, max_iter)

    return mandelbrot_set

xmin, xmax, ymin, ymax = -2, 2, -2, 2
mandelbrot_set = generate_mandelbrot_set(xmin, xmax, ymin, ymax, nx=800, ny=800, max_iter=100)

plt.imshow(mandelbrot_set.T, extent=[xmin, xmax, ymin, ymax], cmap='magma')
plt.xlabel('Re(c)')
plt.ylabel('Im(c)')
plt.title('Mandelbrot Set')
plt.show()
```
