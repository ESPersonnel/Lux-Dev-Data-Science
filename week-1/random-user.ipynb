```python
import requests

def get_top_100_male_users():
    # Make a request to the API to get a random user
    response = requests.get("https://randomuser.me/api/?results=100&gender=male")

    # If the API returns a successful response
    if response.status_code == 200:
        # Get the JSON data from the response
        data = response.json()

        # Get the list of users
        users = data.get("results")

        # Iterate over the users and print their first name and last name
        for i, user in enumerate(users):
            first_name = user.get("name").get("first")
            last_name = user.get("name").get("last")
            print(f"{i + 1}. {first_name} {last_name}")
    else:
        # If the API returns an unsuccessful response, print an error message
        print("Failed to retrieve data from API")

if __name__ == '__main__':
    get_top_100_male_users()
```
