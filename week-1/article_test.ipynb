```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
boston = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
```

```python
boston.head()
```

```python
boston.info()
```

```python
boston.describe()
```

```python
sns.histplot(boston['medv'])
```

```python
sns.scatterplot(x='crim', y='medv', data=boston)
```

```python
# Create a scatter plot of ptratio vs. crim, colored by medv
plt.scatter(boston['ptratio'], boston['crim'], c=boston['medv'])
plt.xlabel('Pupil-teacher ratio')
plt.ylabel('Per capita crime rate')
plt.colorbar()
plt.show()

# Calculate the correlation coefficients
print(boston[['ptratio', 'crim', 'medv']].corr())
```

```python
# Create a scatter plot of lstat vs. medv
plt.scatter(boston['lstat'], boston['medv'])
plt.xlabel('Percentage of lower status population')
plt.ylabel('Median home value')
plt.show()

# Calculate the correlation coefficient
print(boston[['lstat', 'medv']].corr())
```

```python
# Create a correlation matrix
corr_matrix = boston.corr()

# Plot the matrix using a heatmap
sns.heatmap(corr_matrix, annot=True)
plt.show()

# Create a bar chart of the distribution of chas
plt.bar(boston['chas'].unique(), boston['chas'].value_counts())
plt.xlabel('Proximity to Charles River')
plt.ylabel('Number of observations')
plt.show()
```
